### ECDF

estimador de funcion de distribucion acumulada

```python
def ecdf(data):
  #Numero de datos
  n = len(data)

  #ordenamos datos
  x = np.sort(data)

  # valores y
  y = np.arange(1,n+1) / n

  return x, y
```

```python
x_vers, y_vers = ecdf(versicolor_petal_lenght)
plt.plot(x_vers, y_vers, marker='.')

plt.xlabel('longitud de petalo')
plt.ylabel('ECDF')

plt.margins(0.02)

plt.show()
```
